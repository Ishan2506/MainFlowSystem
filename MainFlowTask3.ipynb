Description:

Responsibility:

    Create a bar chart and a line chart using Matplotlib to visualize data
    from a Pandas DataFrame. Customize the charts with labels, titles, and legends.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/householdtask3.csv')
df.head()
```

```python
# Bar chart: Total households vs Owners
plt.figure(figsize=(10, 5))
plt.bar(df['year'], df['tot_hhs'], width=0.4, label='Total Households', align='center')
plt.bar(df['year'] + 0.4, df['own'], width=0.4, label='Owners', align='center')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Total Households and Owners')
plt.legend()
plt.show()
```

```python
# Line chart: Income vs Expenditure
plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['income'], marker='o', label='Income')
plt.plot(df['year'], df['expenditure'], marker='o', label='Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Income and Expenditure Over Years')
plt.legend()
plt.show()
```

```python
# Bar chart
years = df['year']
total_households = df['tot_hhs']
own_wm_prop = df['own_wm_prop']
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(years, total_households, color='green')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Households by Year')
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.plot(years, own_wm_prop, marker='o', linestyle='-', color='purple')
plt.xlabel('Year')
plt.ylabel('Proportion (%)')
plt.title('Proportion of Households Owning Washing Machines by Year')
plt.xticks(rotation=45)

# Show the plots
plt.tight_layout()
```
